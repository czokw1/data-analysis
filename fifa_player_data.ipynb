```python
data = pd.read_csv('data/fifa_data.csv')
data.drop(columns=['Unnamed: 0', 'ID', 'Photo', 'Flag', 'Club Logo'], axis='columns', inplace=True)
#pd.options.display.max_columns = 90 
data.sample(5)
```

## GROUPING & PLOTTING

```python
team = data.groupby('Club')
len(team)
```

```python
team.size().nlargest(5)
```

```python
arsenal = team.get_group('Arsenal')

arsenal = pd.DataFrame(arsenal)
```

```python
arsenal['Wage'] = arsenal['Wage'].str.replace('€','').str.replace('K','')
```
